## **5-What is the average family income of the student?**

```python
def extract_numeric_income(income_str):
    parts = income_str.split()
    numeric_part = parts[0]
    try:
        numeric_income = float(numeric_part)
    except ValueError:
        numeric_income = np.nan  # Set to NaN for now
    return numeric_income

# Apply the function to create a new column 'Numeric Family Income'
df['Numeric Family Income'] = df['Family Income'].apply(extract_numeric_income)

# Compute the average family income
average_family_income = df['Numeric Family Income'].mean()
print(f"The average family income of the students is {average_family_income:.2f} Lakhs.")






```

## **6- How does the GPA vary among different colleges? (Show top 5 results only)**

```python
gpa_column = 'CGPA'
college_gpa = df.groupby('College Name')[gpa_column].mean().reset_index()
top_5_colleges = college_gpa.sort_values(by=gpa_column, ascending=False).head(5)
plt.figure(figsize=(12, 6))
sns.barplot(x=gpa_column, y='College Name', data=top_5_colleges, palette='viridis')
plt.title('Top 5 Colleges by Mean GPA')
plt.xlabel('Mean GPA')
plt.ylabel('College Name')
plt.tight_layout()
plt.show()
```

## **7-Are there any outliers in the quantity (number of courses completed) attribute?**

```python
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['Quantity'] < lower_bound) | (df['Quantity'] > upper_bound)].sum()

# Check if there are any outliers
if len(outliers) > 0:
    print("7. There are outliers in the 'Quantity' (number of courses completed) attribute.")
    print("Outliers:")
    print(outliers)
else:
    print(" There are no outliers in the 'Quantity' (number of courses completed) attribute.")





```

## **11-How does the expected salary vary based on factors like 'GPA', 'Family income', 'Experience with Python(Month)'?**
